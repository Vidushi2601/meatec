```python
import pandas as pd

def parse_time(t):
    t = str(t).strip()
    
    # Case 1: contains days like "1-08:09:45"
    if "-" in t:
        day_part, time_part = t.split("-")
        days = int(day_part)
        h, m, s = map(int, time_part.split(":"))
        total_seconds = days*86400 + h*3600 + m*60 + s
        return total_seconds
    
    # Case 2: standard HH:MM:SS
    else:
        h, m, s = map(int, t.split(":"))
        total_seconds = h*3600 + m*60 + s
        return total_seconds

# Apply to your dataframe
df_final["Test_Time_seconds"] = df_final["Test_Time(s)"].apply(parse_time)

```

```python
df3["dCapacity"] = df3["Current(A)"] * (df3["dt"] / 3600)
df3["dCapacity"] = df3["Current(A)"] * (df3["dt"] / 3600)
df3.groupby("Cycle_Index")["dCapacity"].sum()
```

```python
cell_capacity = 2.3  # Ah from manufacturer

df3["C_rate"] = df3["Current(A)"].abs() / cell_capacity
```

```python
df3["Charge_C_rate"] = df3["Current(A)"].apply(
    lambda x: x/cell_capacity if x > 0 else 0
)

df3["Discharge_C_rate"] = df3["Current(A)"].apply(
    lambda x: abs(x)/cell_capacity if x < 0 else 0
)
```

```python
#make sure dCapacity is correct for every step:
df3["dt"] = df3.groupby("Cycle_Index")["Test_Time_seconds"].diff().fillna(0)
df3["dt"] = df3["dt"].clip(lower=0)

df3["dCapacity"] = df3["Current(A)"] * df3["dt"] / 3600

#now define your function. Discharge capacity should be positive.
def compute_cycle_capacity(cycle_df):
    charge = cycle_df.loc[cycle_df["Current(A)"] > 0, "dCapacity"].sum()
    discharge = cycle_df.loc[cycle_df["Current(A)"] < 0, "dCapacity"].sum()
    # Discharge current is negative, so take abs value
    discharge = abs(discharge)
    return pd.Series([charge, discharge], index=["Charge_Capacity_Ah", "Discharge_Capacity_Ah"])


cycle_capacity = df3.groupby("Cycle_Index").apply(compute_cycle_capacity).reset_index()


```

```python
reference_capacity = cycle_capacity[cycle_capacity["Cycle_Index"] == 1]["Discharge_Capacity_Ah"].iloc[0]

#soh calculation
cycle_capacity["SoH"] = (cycle_capacity["Discharge_Capacity_Ah"] / reference_capacity) * 100
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Round C-rate to group similar values (example: 0.95 -> 1.0)
capacity_crate["C_rate_group"] = capacity_crate["C_rate"].round(1)

unique_crates = capacity_crate["C_rate_group"].unique()
unique_crates
```

```python
degree = 2  

plt.figure(figsize=(10, 6))

for crate in unique_crates:
    df_group = capacity_crate[capacity_crate["C_rate_group"] == crate]
    
    X = df_group["Cycle_Index"].values.reshape(-1,1)
    y = df_group["Charge_Capacity_Ah"].values
    
    # Polynomial transformation
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    
    # Fit regression model
    model = LinearRegression()
    model.fit(X_poly, y)
    
    # Extrapolate across cycles
    cycle_range = np.linspace(df_group["Cycle_Index"].min(),
                              df_group["Cycle_Index"].max() + 20, 200).reshape(-1,1)
    cycle_poly = poly.transform(cycle_range)
    y_pred = model.predict(cycle_poly)
    
    # Plot actual points
    plt.scatter(X, y, s=25, label=f"Actual (C={crate})")
    
    # Plot predicted polynomial curve
    plt.plot(cycle_range, y_pred, label=f"Predicted (C={crate})")

plt.title("Polynomial Regression: Charge Capacity vs Cycle Count (Per C-rate)")
plt.xlabel("Cycle Count")
plt.ylabel("Charge Capacity (Ah)")
plt.grid(True)
plt.legend()
plt.show()
```
